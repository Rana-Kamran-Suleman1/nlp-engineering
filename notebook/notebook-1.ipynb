```python
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer       # Streams
from nltk.stem import WordNetLemmatizer
```

```python
# Sample text
text = "This is a sample sentence showing stopword removal."

# Get English stopwords and tokenize
stop_words = set(stopwords.words('english'))
tokens = word_tokenize(text.lower())

# Remove stopwords
filtered_tokens = [word for word in tokens if word not in stop_words]

print("Original:", tokens)
print("Filtered:", filtered_tokens)
print(stop_words)
```

```python
stemmer=PorterStemmer()
print(stemmer.stem("Happiness"))
```

```python
lemmatizer=WordNetLemmatizer()
print(lemmatizer.lemmatize('Happiness'))
print(lemmatizer.lemmatize("Congratulations"))
```
